```python
def inss(v):
    for i in range(1,len(v)):
        j=i;
        while j>0 and v[j]<v[j-1]:
            v[j],v[j-1]=v[j-1],v[j];
            j=j-1;
    return(v)
```

```python
def bub(v):
    for i in range(len(v)):
        for j in range(0, len(v)-i-1):
            if v[j] > v[j+1] :
                v[j],v[j+1] = v[j+1],v[j]
    return(v)
```

```python
def merg(v):
    if len(v) > 1:
        mid = len(v)//2;
        L = v[:mid];
        R = v[mid:];
        merg(L);
        merg(R);
        i = j = k = 0;
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                v[k] = L[i];
                i += 1;
            else:
                v[k] = R[j];
                j += 1;
            k += 1
        while i < len(L):
            v[k] = L[i];
            i += 1;
            k += 1;
        while j < len(R):
            v[k] = R[j];
            j += 1;
            k += 1;
    return(v)
```

```python
def heapify(v,n,i):
    larg = i;
    l = 2 * i + 1; 
    r = 2 * i + 2;    
 
    if l < n and v[larg] < v[l]:
        larg = l;
 
    if r < n and v[larg] < v[r]:
        larg = r;
 
    if larg != i:
        v[i], v[larg] = v[larg], v[i] 
 
        heapify(v, n, larg)
```

```python
def heaps(v):
    n=len(v);
    for i in range(n//2 - 1, -1, -1):
        heapify(v, n, i)
    for i in range(n-1, 0, -1):
        v[i], v[0] = v[0], v[i]
        heapify(v, i, 0)
    return(v)
```

```python
def partition(v, l, h):
    i = (l-1)  
    pivot = v[h]   
 
    for j in range(l, h):
        if v[j] <= pivot:
            i = i+1
            v[i], v[j] = v[j], v[i]
 
    v[i+1], v[h] = v[h], v[i+1]
    return (i+1)
```

```python
def quicks(v, l, h):
    if len(v) == 1:
        return v
    if l < h:
        pi = partition(v, l, h)

        quicks(v, l, pi-1)
        quicks(v, pi+1, h)
    return(v)
```

```python
import random
def rlis(n):
    v=[]
    for i in range(n):
        v.append(random.randint(-10000, 10000))
    return(v)
```

```python
import random
def inlis(n):
    v=[]
    k=int(-10000)
    for i in range(n):
        s=random.randint(k, k+10000)
        v.insert(0,s)
        k=s
    return(v)
```
